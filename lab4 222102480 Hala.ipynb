```python
import numpy as np
```

```python
def pearson_coef(user1,user2):

    mask = []
    for i in range(len(user1)):
        if user1[i] != 0 and user2[i] != 0:
            mask.append(i)
            
    if len(mask) == 0:
        return 0
    
    u1 = [user1[i] for i in mask]
    u2 = [user2[i] for i in mask]

    sum1 = 0
    sum2 = 0

    for i in range(len(u1)):
        sum1 += u1[i]
        sum2 += u2[i]

    mean1 = sum1/ len(u1)
    mean2 = sum2/ len(u2)

    dev1 =[]
    dev2 =[]

    for i in range (len(u1)):
        dev1.append(u1[i] - mean1)
        dev2.append(u2[i] - mean2)

    num = 0
    for i in range(len(dev1)):
        num += dev1[i] * dev2[i]
        
    sum_sq1= 0   
    sum_sq2= 0

    for i in range(len(dev1)):
        sum_sq1 += dev1[i] **2
        sum_sq2 += dev2[i] **2
    
    denom = (sum_sq1 * sum_sq2 ) **0.5
    if denom ==0:
        return 0
    
    return num/denom

```

```python
usera = np.array([5,3,0,4,0,2])
userc = np.array([4,3,4,3,5,3])

corr = pearson_coef(usera, userc)
print(f"Pearson Correlation:{corr:.3f}")
```
